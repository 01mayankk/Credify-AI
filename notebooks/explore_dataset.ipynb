```python
#---------------------
#STEP 1 : Import Libraries
#---------------------
import numpy as np
import pandas as pd
```

```python
#---------------------------------
# STEP 2:  Load the Dataset
#---------------------------------

#Load the CSV file
df = pd.read_csv("../data/Credify_Dataset_1.csv", low_memory = False)

#We used low_memory = False because the dataset is large and to avoid dtype warning
#"../data/..." means : go up one foldr, then go to data folder, then access the file

# Show shape of dataset(rows, columns)
df.shape
```

```python
#---------------------------------
# STEP 5:  Check distribution of loan_status
#---------------------------------
#Loan status tells us how the loan ended
#We will later convert it to Low/Moderate/High risk
df['loan_status'].value_counts(dropna = False)
```

```python
#---------------------------------
# STEP 6:  Check missing values per columns
#---------------------------------

missing_count = df.isna().sum() # How many missing values per column
missing_percent = (df.isna(). mean() * 100).round(2) # Percentage of missing values per column

# Combine oth into one table
missing_table = pd.DataFrame({
    "missing_values" : missing_count,
    "missing_percentage" : missing_percent
})

# Sort by heighest missing values
missing_table.sort_values(by = "missing_values", ascending = False).head(20)
```
